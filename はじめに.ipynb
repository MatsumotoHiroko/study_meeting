# 5. 機械学習のメリット・デメリット
---
## 機械学習とは
### 機械に人間のような判断をさせる
機械学習は古くからあるAIの手法の1つで、機械に学習させて人間と同じような認識、判断を行わせるもの

[参考: 第1回 ビジネスに活用するためのAIを学ぶ](https://thinkit.co.jp/article/12744)

---
## 機械学習のメリット
### 大量データを解析し予測する
![機械学習のメリット](https://markezine.jp/static/images/article/26305/26305_006.png)

メリットをまとめると、「大量のデータから未来を予測したり、あらたな法則を見出しビジネスに活かす。データが増えるほど精度が向上する」

[参考: MA活用のポイントは、機械学習と人力の使い分けにあり](https://markezine.jp/article/detail/26305?p=3)

---
## 機械学習のデメリット
### 制約や前例がない問いには不向き
![機械学習のデメリット](https://markezine.jp/static/images/article/26305/26305_007.png)

「制約なし・前例なし・答えが複数・過去のデータに傾向なし」対応が難しい
「あらかじめトリガーが何かわかっている場合」　仮説に基づいて施策を実行すればいい（機械学習不要）
「モデル式を作るのに時間がかかる」いまは計算のためのインフラが整っているので改善されている

[参考: MA活用のポイントは、機械学習と人力の使い分けにあり](https://markezine.jp/article/detail/26305?p=3)


# 6. 機械学習の手法
---
1. 回帰(Regression)
2. 分類(Classification)
3. クラスタリング(Clustering)



## 6-1. 回帰
---
### 連続データから数値を予測

* 連続するデータ（ex時系列データ）を別の連続するデータから予測
* 相関関係を見る

など

#### 時系列データ、相関関係
![身長から体重を予測](https://www.albert2005.co.jp/knowledge/images/tech_multivariate_img19.jpg)

[参考:単回帰分析とは](https://www.albert2005.co.jp/knowledge/statistics_analysis/multivariate_analysis/single_regression)


#### 連続データから予測
* 薬局の数と人口密度
![散布図](http://bellcurve.jp/statistics/wp-body/wp-content/uploads/2017/01/795316b92fc766b0181f6fef074f03fa.png)
![xx次関数:人口密度](http://bellcurve.jp/statistics/wp-body/wp-content/uploads/2017/01/2b530e80c7d0de90885e285c5d798063.png)

式(モデル)を変えることによって、単純な直線だけでなく曲線での予測もできる

[参考：27-1. 単回帰分析](https://bellcurve.jp/statistics/course/9700.html)



## 6-2. 分類
---
### データを決められたカテゴリに振り分ける
* 迷惑メールか、それ以外か

![メール振り分け](http://www.nifty.com/mail/antispam/img/spamtaisaku_img05.jpg)

[参考:nifty迷惑メールフィルター](http://www.nifty.com/mail/antispam/filter.htm)

* 手書き数字がどの数字か
![MNIST](https://www.wolfram.com/language/11/neural-networks/assets.ja/digit-classification/smallthumb_1.png)
[参考;オブジェクトの分類](https://www.wolfram.com/language/11/neural-networks/object-classification.html?product=language)

最初から答えを与えて（この場合、「迷惑メールである・ない」、「数字0-1」）学習させる（教師あり学習）

[参考：機械学習 はじめよう 第15回　分類問題ことはじめ](http://gihyo.jp/dev/serial/01/machine-learning/0015)

---
### 補足：教師あり学習

* 先生に従って学習するように、正解がわかっている場合の学習のこと
* 問題とその正解のペア（学習データ）が与えられた状態で学習
* 問題を解き、結果を正解と比較、正解との誤差を小さくするようパラメータ更新する

<img src="image/supervised.png">
[参考:おうちで学べる人工知能のきほん 5-1-1 機械学習の種類を知ろう]

## 6-3. クラスタリング
---
### データを似た特徴でまとめてグループ化
バラバラなデータから似たような特徴をまとめていくつかのグループに分ける
* 顧客のセグメンテーション（性別、地域、ライフスタイル、購入履歴などの類似でグループに分ける）
![顧客](http://bdm.change-jp.com/wp-content/uploads/2014/11/cap04.png)

[参考:クラスター分析「使ってみたくなる統計」シリーズ 第3回](http://bdm.change-jp.com/?p=2025)

答えのないデータで学習させる（教師なし学習）

---
### 補足：教師なし学習

* 正解がない場合の学習のこと
* 問題を観察し、類似点のまとまりを発見する
* 未知の問題に対し、どのまとまりに対する問題か認識する

<img src="image/unsupervised.png">

[参考:おうちで学べる人工知能のきほん 5-1-1 機械学習の種類を知ろう]
