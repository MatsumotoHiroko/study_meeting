# 微分
## 瞬間の変化量を求めるのが微分
微分を使うことで、関数のある点における「傾き・瞬間の変化量」を捉えることができる。

車で街を走ることを考える。
横軸（x）を経過時間、縦軸（y）を走行距離とすると、以下のようなグラフに表せられる。

![全体の速度](image/bibun1.png)

このグラフにより、

1. 40秒で120mほど走行しているため、全体で $ \frac{120m}{40s} = 3m / s $ の速度が出ていたと考えられる。
2. 1は平均速度のため、常に3m / sではない。発車時はゆっくり走り、信号停止では速度0になり進まない。

「ある時点におけるその瞬間の速度」はそれぞれ異なるといったことがわかる。

* 道のり(距離) 速さ 時間の公式を忘れた方のために

![道のり速さ時間公式](http://hensa40.cutegirl.jp/wp-content/uploads/2015/12/hayasa.png)

1. 道のり(距離)(km) = 速さ(km/h) × 時間(h)
2. 速さ(km/h) = 道のり(距離)(km) ÷ 時間(h)
3. 時間(h) = 道のり(距離)(km) ÷ 速さ(km/h)

[参考：[ 小中学生 ] 道のり 速さ 時間の公式を理解しよう](http://hensa40.cutegirl.jp/archives/375)

40秒間全体ではなく、「瞬間の変化量」を求めるため、だんだん間隔を狭めて考える。
以下のように、10秒から20秒に着目すると、その間では約60mほど走行している。

![範囲の速度](image/bibun2.png)

この間の速度は、 $ \frac{60m}{10s} = 6m / s $ と求められる。

これは、[一次関数の式](#一次関数の式)で説明した、「xが1増えるとyがいくつ増えるか」の、グラフの傾きを求めることと等しい。
今回は10秒間に注目したが、これを「10.0秒と10.1秒の間」という具合でどんどん間隔を縮めると１０秒時点のその瞬間の変化量（傾き、速度）がわかる。

![範囲の速度](image/bibun3.png)

## 瞬間の変化量を求める式
「瞬間の変化量（傾き）」を求める関数を $ f(x) $と置き、hを限りなく間隔を縮めて0に近づけた数字とすると、$ f(x) $の瞬間のxは以下のような式で示すことができる。

 $ \frac{d}{dx}f(x) = \lim_{h \to 0} \frac{f(x+h)-f(x)}{h} $
 
*  $ f(x) $ 瞬間の変化量を求める関数
* h 限りなく間隔を縮めて0に近づけた数字。「h→0」は「hを限りなく0に近づける」という意味である。

具体的に「10.0秒と10.1秒の間」の傾きを考えた場合、

* $ x = 10.0 $
* $ h = 10.1 $

と考えられる。

仮に、

* 10.0秒の時点　→ 40.0mの速度で走行
* 10.1秒の時点　→ 40.6mの速度で走行

とすると、以下の式で求められる。

$ \frac{f(10 + 0.1)-f(10)}{0.1} = \frac{40.6 - 40}{0.1} = 6 $

この6が傾きであり、今回のケースでは速度に当たる。
（本来はhは限りなく0に近づける必要があるが例としてh=0.1で説明）

以上、関数$ f(x) $の点xにおける傾きを求めることを「微分する」という

## 微分を表す表現
全て「関数を微分する」という意味で、表記の違いだけである。

$ f(x) $の微分を表現する場合、微分演算子、プライム を使った方法がある。

### 微分演算子  

1. $ \frac{df(x)}{dx} $
2. $ \frac{d}{dx}f(x) $

### プライム 

1. $ f'(x) $

## 導関数の性質
### 微分の公式
微分は[瞬間の変化量を求める式](#瞬間の変化量を求める式)で説明した　 $ \frac{d}{dx}f(x) = \lim_{h \to 0} \frac{f(x+h)-f(x)}{h} $の式でも求められるが、以下の公式でも求められる。

$ f(x) = x^n $を微分すると、
$ \frac{d}{dx}f(x)　= nx^n-1 $

### 定数倍、和（差）
* 関数$ f(x) $、$ g(x) $を足す（引く）場合、$ f(x) $、$ g(x) $をそれぞれ微分して足す（引く）
$ \frac{d}{dx}(f(x) ± g(x)) =\frac{d}{dx}f(x) ± \frac{d}{dx}g(x) $

* 定数a(関数ではない値）を関数$ f(x) $に掛けている場合、関数$ f(x) $を微分した結果にそのまま掛ける
$ \frac{d}{dx}(af(x))　=  a\frac{d}{dx}(f(x))$

### 定数項
* 微分対象であるxに関係のない定数aの微分は0
$ \frac{d}{dx}a = 0 $

# 偏微分
多変数関数を，特定の文字以外定数だとみなして微分したものを偏微分と呼ぶ。

x に関する偏微分：x 以外定数だと思って微分。 $ \dfrac{\partial f}{\partial x}(x,y),\:f_x $ などと表す。

y に関する偏微分：y 以外定数だと思って微分。 $ \dfrac{\partial f}{\partial y}(x,y),\:f_y $ などと表す。

# ２次関数
## 展開公式

\begin{equation*}
(x+a)(x+b)=x^2+(a+b)x+ab\\ 
(a+b)(a-b)=a^2-b^2\\ 
(a+b)^2=a^2+2ab+b^2\\ 
(a-b)^2=a^2-2ab+b^2\\ 
(ax+b)(cx+d)=acx^2+(ad+bc)x+bd
\end{equation*}

# 機械学習での一次関数の表現
y = ax + bの式は、変形して、

$ f_{\theta}(x) = {\theta}_0 + {\theta}_1x $

と表す。
意味は、この関数がθというパラメータを持っていて、かつxという変数の関数だ
ということである。

+ y = $ f_{\theta}(x) $。  $ f $はfunction（関数）の頭文字をとったもの。 
e.g. y = x + 7において、x=2のときのyの値は？と聞かれた場合、 y = $ f $とおいて、$ f(2) = 2 + 7 = 9 $ と示したりする。

+ θ これから求めていく未知数、パラメータのこと。 今回は、 $ {\theta}_0 = b $、 $ {\theta}_1 = a $ 

# 誤差の表現
「誤差」は、式を変形すると
$ y - f_{\theta}(x) $ (データy:クリック数 - 一次関数にyに対応するx:広告費を入れたときの算出結果)　で表せる。

誤差がないというのは、誤差が0ということなので、
$ y - f_{\theta}(x) = 0 $
の状態が理想的。


# 二乗誤差関数
学習データ毎の誤差の和（誤差を全部足した値）は次の式で示すことができる。

$ E({\theta}) = \frac{1}{2}\sum_{i=1}^{n} ({y^{(i)} - f{\theta}(x^{(i)})})^2 $

この　$ E({\theta})　$ の値を一番小さくする$ {\theta} $を見つけることが、関数のグラフをデータ上にプロットすることとなる。

* Eは、誤差の英訳でErrorの頭文字
* $ \sum $ は総和の記号。　
*　(i) は、何番目のデータかを示すもの
* $  \frac{1}{2} $ を掛けているのは、あとで微分をする際に式を簡単にするため。

# 総和（Σ）
$ \sum $　は、足し算を表す時に便利。
1 + 2 + 3 + ・・・ + 99 + 100 をΣを使って表すと、 
$\displaystyle \sum_{n=1}^{100} i $

# 最適化問題
「最適化問題」：一番小さい値を見つけることを目的とすること
特徴：$ \frac{1}{2} $のような勝手な定数を掛けても問題ない。正の定数をかけても「どこが最小値になるか」は変わらないから。


例えば、以下のような$ f_{\theta}(x) = x^2 $のようなグラフがあったとする。これが最小になるxは$  x = 0 $である。

```python
x = np.linspace(-3,3,20) # 0～3まで20刻みでxの値を生成
y = x**2            # 曲線の式(2次関数)
plt.plot(x,y,"r-")      # 曲線を引く
plt.show()              # グラフ表示
```

上のグラフに$ \frac{1}{2} $を掛けた$ f_{\theta}(x) = \frac{1}{2}x^2 $というグラフが最小になるxも$  x = 0 $である。

```python
x = np.linspace(-3,3,20) # 0～3まで20刻みでxの値を生成
y = 1 / 2 * x**2            # 曲線の式(2次関数)
plt.plot(x,y,"r-")      # 曲線を引く
plt.show()              # グラフ表示
```

# 最急降下法
$ E({\theta})　$（全体の誤差の合計）を小さくするために、$ {\theta}　$の値を適当に変えながら$ E({\theta})　$を計算し、前の値と比較するのはとても面倒なので、「微分」を使って求めてゆく。
※微分については、[数学説明.ipynb#微分](数学説明.ipynb#微分)を参照


## 増減表
微分は変化の度合いを求めるために使うものである。
ここで、まず、「増減表」（関数のグラフのマイナス、プラスの符号が変わる範囲をまとめた表）を作成し、関数から求められる値を小さくしていくやり方を具体的に見ていく。

例として、$ g(x) = (x - 1)^2 $という二次関数で考える。
$ g(x) = (x - 1)^2 $　の最小値は$ x = 1$の際の$ g(x) = 0 $ (xに0を代入すると　$ (1 - 1)^2 = 0^2 = 0 $
<img src="image/g_x_min.png">

$ g(x) $を微分する。
式を展開すると$ (x - 1)^2 = x^2 - 2x + 1 $
[式の展開](数学説明.ipynb#２次関数)

微分すると

$ \frac{d}{dx}g(x) = 2x - 2 $

増減表は導関数（微分した後の関数のこと。この場合$ 2x - 2 $）の符号に着目

| xの範囲 | $ \frac{d}{dx}g(x)の符号 $ | g(x)の増減 |
|:-----------|------------:|:------------:|
| x < 1 | -  | ↘︎  
| x = 1 | 0  |   |
| x > 1 | +  | ↗︎  |
<img src="image/2x_2_plamai.png">

$ x = 3 $ からスタートして$ g(x) $の値を小さくするにはxを減らしていく（左にずらす）
<img src="image/x3start.png">

$ x = -1 $からスタートして$ g(x) $の値を小さくするにはxを増やしていく（右にずらす）
<img src="image/m1start.png">

導関数の符号と逆の方向にずらせば最小値の方に動いてパラメータが更新される仕組み。

## 最急降下法の式
以上を「最急降下法（勾配降下法）」といい、次の式で表す

$ x := x - η\frac{d}{dx}g(x) $
（A := B はAをBによって定義する、の意味）
この式の場合、「新しいxを１つ前の古いxを使って定義する」（xを最急降下法の式で算出された結果で上書き）という意味になる

### 学習率 η
η（イータ）は学習率と呼ばれる、整の定数
学習率の大小により最小値にたどり着くまでの更新回数が変わってくる。

学習率が大きすぎると次のように同じ数値を行ったり来たりしてループしてしまう

* $ η = 1 $の場合
<img src="image/2x_2loop.png">

学習率が小さすぎると次のようにだんだん最小値に近づいていくものの、移動量が小さくてなかなか更新されない

* $ η = 0.1 $の場合
<img src="image/2x_2_m1start.png">

## 最終的なパラメータθの更新式
パラメータ更新式（最急降下法）

$ x := x - η\frac{d}{dx}g(x) $

を、今回の目的関数・予測関数で必要とされるパラメータにそれぞれ当てはめる。

* 目的関数(最小二乗法)
$ g(x) = E({\theta}) = \frac{1}{2}\sum_{i=1}^{n} ({y^{(i)} - f{\theta}(x^{(i)})})^2 $

* パラメータ(予測関数の $ {\theta}_0 + {\theta}_1x $)
$ x = {\theta}_0,  {\theta}_1 $

そして、$  {\theta}_0,  {\theta}_1 $ それぞれのパラメータでパラメータ更新式を偏微分する
[偏微分について](数学説明.ipynb#偏微分)

$ {\theta}_0 := {\theta}_0 - η\frac{\partial E}{\partial \theta_0} $

$ {\theta}_1 := {\theta}_1 - η\frac{\partial E}{\partial \theta_1} $

偏微分の結果、以下が最終的な$  {\theta}_0,  {\theta}_1 $ それぞれのパラメータの更新式になる。

$ {\theta}_0 := {\theta}_0　- η\sum_{i=1}^{n} (f{\theta}(x^{(i)}) - y^{(i)}) $

$ {\theta}_1 := {\theta}_1　- η\sum_{i=1}^{n} (f{\theta}(x^{(i)}) - y^{(i)}) $

この式に従って$  {\theta}_0,  {\theta}_1 $を更新することで、誤差を最小限にした予測関数が見つかる仕組み
